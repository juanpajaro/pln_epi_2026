```python
import spacy
import medspacy
```

```python
with open("./texto_historia_v2.txt") as f:
    text = f.read()
```

```python
nlp = spacy.blank("es")
nlp.pipe_names
```

Tokenization

```python
from medspacy.custom_tokenizer import create_medspacy_tokenizer
```

```python
medspacy_tokenizer = create_medspacy_tokenizer(nlp)
default_tokenizer = nlp.tokenizer
```

```python
example_text = r'Pt c\o n;v;d h\o chf+cp'
```

```python
print("Tokens from default tokenizer:")
print(list(default_tokenizer(example_text)))
print()
print("Tokens from medspacy tokenizer:")
print(list(medspacy_tokenizer(example_text)))
```

```python
nlp.tokenizer = medspacy_tokenizer
print(list(nlp(example_text)))
```

Segmentacion de oraciones

```python
#PyRush es un algoritmo de segmentación de oraciones diseñado para manejar el lenguaje médico, 
# que a menudo contiene abreviaturas, símbolos y estructuras de oración complejas. 
# PyRuSH es una implementación en Python de este algoritmo, que se integra fácilmente con spaCy y 
# medspacy para mejorar la precisión de la segmentación de oraciones en textos médicos.
from medspacy.sentence_splitting import PyRuSHSentencizer
```

```python
nlp.add_pipe("medspacy_pyrush") 
```

```python
nlp.pipe_names
```

```python
doc = nlp(text)
```

```python
def cargar_historia(ruta_archivo):
    """
    Carga el contenido de un archivo de texto con la historia clínica.
    
    Args:
        ruta_archivo (str): Ruta al archivo de texto que contiene la historia clínica
        
    Returns:
        str: Contenido del archivo
        
    Raises:
        FileNotFoundError: Si el archivo no existe
        IOError: Si hay un error al leer el archivo
    """
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
        return contenido
    except FileNotFoundError:
        print(f"Error: El archivo '{ruta_archivo}' no fue encontrado.")
        raise
    except IOError as e:
        print(f"Error al leer el archivo: {e}")
        raise
```

```python
# Ejemplo de uso:
historia = cargar_historia('texto_historia_v2.txt')
doc = nlp(historia)

for ent in doc.ents:
    print('Entity text : {}'.format(ent.text))
    print('Label (UMLS CUI) : {}'.format(ent.label_))
    print('Similarity : {}'.format(ent._.similarity))
    print('Semtypes : {}'.format(ent._.semtypes))
```
